# 1)  Modellere Myoglobin 

Vi starter med å se på proteinet *myoglobin*. Dette er et protein vi finner i muskelvev hos alle virveldyr, inkludert mennesker. Myoglobin kan binde til seg oksygen, og virker på den måten som et slags okysgenlager i musklene.

### Modellere binding av oksygengass til myoglobin

Vi starter med å se på hvordan oksygen binder seg til myoglobin. 

Et myoglobinprotein har en form der den har et bestemt omeråde hvor et okysgengass atom kan binde seg fast. Vi skriver reaksjonen at oksygen binder seg til myoglobinet som følger
$$\rm Mb + O_2 \rightarrow MbO_2.$$
Her er "Mb" et myoglobin-protein uten oksygen tilknyttet seg, O$_2$ er et oksygengassmolekyl og MbO$_2$ er et *oksymyoglobin* molekyl, altså et myoglobin-protein med et okysgengass molekyl festet til seg.

Vi ønsker nå å *modellere* hvordan denne reaksjonen utspiller seg over tid. Med andre ord ønsker vi å vite hvordan konsentrasjonen av de tre molekylene endrer seg over tid. For å gjøre dette formulerer vi hvordan endringene av hver konsentrasjon er.

Først ser vi på endringen i konsentrasjon av myoglobin: $\rm [Mb]$. Etterhvert som reaksjonen utspiller seg vil det bli mindre og mindre myoglobin. Jo flere reaksjoner som skjer, jo raskere faller konsentrasjonen. For at en reaksjon skal finne sted må et myoglobin-protein og et oksygen-atom møtes. Sannsynligheten for det er større jo høyere konsentrasjonen er av disse molekylene, derfor må hastigheten på reaksjonen være *proporsjonal* med disse konsentrasjonene. Vi kan skrive dette ut som ligningen:
$$[{\rm Mb}]' = -k[{\rm Mb}][{\rm O_2}].$$
Her er $[{\rm Mb}]'$ den *deriverte* av $\rm [Mb]$, som sier hvordan konsentrasjonen endrer seg over tid. Vi har også innført $k$, som er ratekonstanten. Dette er en parameter som er forskjellig for forskjellige reaksjoner og som vi måler eksperimentelt. Den jo større ratekonstanten er, jo raskere går reaksjonen. Ratekonstanten vil også øke med temperatur og kan også variere med faktorer som trykk, pH, osv.

Ettersom at reaksjonens hastighet beskrives av ratekonstanten $k$ er det vanlig å skrive opp selve reaksjonen som:
$$\rm Mb + O_2 \overset{k}{\rightarrow} MbO_2.$$



Endringen i konsentrasjon av oksygen vil være helt lik som for myoglobin, fordi reaksjonen er mellom nøyaktig ett molekyl av hver type. Dette kalles det støkiometriske forholdet.
$$[{\rm O_2}]' = -k[{\rm Mb}][{\rm O_2}].$$

Ettersom at det dannes nøyaktig ett oksymyoglobin-molekyl for hver reaksjon som skjer, så vil reaksjonsraten til oksymyoglobin være lik som de to andre konsentrasjonene. Derimot *dannes* det oksymyoglobin, så konsentrasjonen øker istedenfor å avta, derfor dropper vi minustegnet:
$$[{\rm MbO_2}]' = k[{\rm Mb}][{\rm O_2}].$$

### Løse Differensialligninger

Vi har altså tre ligninger som beskriver reaksjonen vår
$$
\begin{align}
[{\rm Mb}]' &= -k[{\rm Mb}][{\rm O_2}], \\
[{\rm O_2}]' &= -k[{\rm Mb}][{\rm O_2}], \\
[{\rm MbO_2}]' &= k[{\rm Mb}][{\rm O_2}].
\end{align}
$$

Dette er et eksempel på tre koblede *differensialligninger*. Differensialligninger er ligninger der den deriverte av en funksjon inngår, og løsningen av slike ligninger er funksjoner. I dette tilfellet har vi tre differensialligninger med tre ukjente. Om vi klarer å løse disse ligningene finne vi funksjonene $[{\rm Mb}](t)$, $[{\rm O}_2](t)$ og $[{\rm MbO}_2](t)$, som beskriver hvordan konsentrasjonene endrer seg over tid.

Å løse koblere differensialligninger av denne typen er ganske utfordrende å gjøre for hånd. Men vi kommer istedet til å løse dem *numerisk*, det vil si ved hjelp av datamaskinen. Da løser vi dem ikke 100% nøyaktig, men vi kommer nært nok til at vi ikke merker forskjell.

## Euler's Metode (Litt utfordrende)

Dette avsnittet dekker hvordan vi løser ligningene matematisk. Det er ment for å gi litt forklaring på hvordan koden vi skal skrive snart fungerer. Derimot er det ikke så viktig for selve kjemien i det hele, så denne biten kan glattes litt over om man ikke bryr seg så mye om matematikken i det hele.

For å løse en differensialligning numerisk bruker vi det som kalles *Euler's metode*. Vi skal nå prøve å forklare baktanken på denne.

Om vi ser på differensialligningen
$$[{\rm Mb}]' = -k[{\rm Mb}][{\rm O_2}],$$
Så er det tre ukjente, som egentlig er funksjoner, så vi kan skrive ut dette mer nøyaktig som
$$[{\rm Mb}]'(t) = -k[{\rm Mb}](t)[{\rm O_2}](t),$$ \\
altså beskriver ligningen hvordan konsentrasjonen endrer seg ved ethvert tidspunkt $t$.

For å løse denne ligningen setter vi inn definisjonen på den deriverte
$$\frac{[{\rm Mb}](t+\Delta t) - [{\rm Mb}](t)}{\Delta t} = -k[{\rm Mb}](t)[{\rm O_2}](t).$$
I matematikken lar man $\Delta t$ "gå mot 0" til den blir uendelig liten. Når man regner *numerisk* lar vi den ikke gå helt til 0, men lar den være et veldig lite tall.

Ettersom at $\Delta t$ bare er et veldig lite tall kan vi gange på begge sider av ligningen, og vi får
$$[{\rm Mb}](t+\Delta t) - [{\rm Mb}](t) = -k[{\rm Mb}](t)[{\rm O_2}](t)\times\Delta t.$$

Og vi kan også flytte over $[{\rm Mb}](t)$, slik at vi har
$$[{\rm Mb}](t+\Delta t) = [{\rm Mb}](t)  -k[{\rm Mb}](t)[{\rm O_2}](t)\times \Delta t.$$

Merk at høyre side av ligningen nå har alle konsentrasjonene ved tiden $t$, mens venstre side har konsentrasjonen ved $t+\Delta t$. Altså lar ligningen regne oss et "tidssteg" $\Delta t$ fremover i tid. Om vi da kjenner konsentrasjonen ved start ($t=0$), dvs, $[\rm Mb](0)$, så kan vi nå regne ut $[\rm Mb](\Delta t)$, så $[\rm Mb](2\Delta t)$, så $[\rm Mb](3\Delta t)$, osv, altså regner vi oss fremover i tid.

For å gjøre notasjonen litt lettere sier vi nå at $t_i = i\cdot \Delta t$:
$$
\begin{align}
t_0 &= 0 \\
t_1 &= \Delta t \\
t_2 &= 2\Delta t \\
    &\ldots
\end{align}
$$

Da kan vi skrive ligningen som
$$[{\rm Mb}](t_{i+1}) = [{\rm Mb}](t_i)  -k[{\rm Mb}](t_i)[{\rm O_2}](t_i)\times \Delta t.$$
For å forenkle enda mer kan vi si at $[{\rm Mb}](t_i)$ skal forkortes som $[{\rm Mb}]_i$, da får vi
$$[{\rm Mb}]_{i+1} = [{\rm Mb}]_i  -k[{\rm Mb}]_i[{\rm O_2}]_i\times \Delta t.$$

Vi kan gjøre det samme for de to andre ligningene, og da finner vi
$$
\begin{align}
[{\rm Mb}]_{i+1} &= [{\rm Mb}]_i  -k[{\rm Mb}]_i[{\rm O_2}]_i\times \Delta t, \\
[{\rm O_2}]_{i+1} &= [{\rm O_2}]_i  -k[{\rm Mb}]_i[{\rm O_2}]_i\times \Delta t, \\
[{\rm MbO_2}]_{i+1} &= [{\rm MbO_2}]_i  + k[{\rm Mb}]_i[{\rm O_2}]_i\times \Delta t.
\end{align}
$$

Dette er kanskje veldig forvirrende, men grunntanken er at vi kjenner konsentrasjonene ved start, og så regner vi ut hvordan de endrer seg over tid ved å ta bitte små steg fremover i tid.



### Simulere reaksjonen

Når vi bruker Euler's metode til å regne oss frem til konsentrasjonene over tid, så kan vi si at vi *simulerer* reaksjonen som skjer. For å få til dette trenger vi en konkret ratekonstant $k$. Vi slår denne opp på nett og finner en verdi av $k = 14$ µm$^{-1}$ s$^{-1}$ ([Kilde](https://www.ncbi.nlm.nih.gov/books/NBK153687/table/T22/)). I tilegg må vi si hva konsentrasjonene er ved start. Vi velger å simulere en situasjon der vi kun har myoglobin og oksygen ved start, og ingen oksymyoglobin. Altså kan vi si
$$\rm [Mb]_0 = 1 {\rm\ µM}, \qquad \rm [O_2]_0 = 3 {\rm\ µM}, \qquad \rm [MbO_2]_0 = 0 {\rm\ µM}.$$

Her sier vi altså at det finnes tre ganger så mye oksygen som myoglobin.

I tilegg må vi velge størrelsen på tidssteget vårt $\Delta t$, og hvor lenge vi skal simulerer totalt. Reaksjonen går raskt, så det holder å simulere $T=1$ sekund. Tidssteget bør være liten, så simuleringen er nøyaktig, vi kan for eksempel velge $\Delta t = 0.001$ sekund.


```python
from pylab import *

T = 1 # Antall sekunder vi skal simulere
dt = 0.001 # Tidssteget, denne må være liten!
N = int(T/dt) # Antall tidssteg vi tar totalt

# Lager tomme arrays å oppbevare løsningene i
t = zeros(N+1)
Mb = zeros(N+1)
O2 = zeros(N+1)
MbO2 = zeros(N+1)

# Parametre
k = 14 # µM⁻¹ s⁻¹

# Initialbetingelser
Mb[0] = 1 # µM
O2[0] = 3 # µM
MbO2[0] = 0 # µM

# Regner oss fremover i tid med Euler's metode
for i in range(N):
    t[i+1] = t[i] + dt
    Mb[i+1] = Mb[i] - k*Mb[i]*O2[i]*dt
    O2[i+1] = O2[i] - k*Mb[i]*O2[i]*dt
    MbO2[i+1] = MbO2[i] + k*Mb[i]*O2[i]*dt
    
# Plotte løsningene
plot(t, Mb, label='Mb')
plot(t, O2, label='O2')
plot(t, MbO2, label='MbO2')

xlabel('Tid (s)')
ylabel('Konsentrasjoner (µM)')
legend()
show()
```

## Modellere en reversibel reaksjon

Så langt har vi kun simulert at oksygen binder seg til myoglobin, men i virkeligheten er dette en *reversibel* reaksjon, altså skjer også den motsatte reaksjonen at oksymyoglobin deler seg opp og frigjør okysgenmolekylet:
$$\rm MbO_2 \rightarrow Mb + O_2.$$
Denne reaksjonen utspiller seg med en annen ratekonstant enn den forrige. For å skille på dem kaller vi ratekonstanten for at oksygen binder seg til myoglobin for $k_{\rm a}$, og ratekonstanten for at de skiller seg igjen for $k_{\rm d}$. Bokstavene kommer fra at molekyler som skiller seg kalles for "dissosiasjon", så $k_{\rm d}$ er *disosiasjonsraten*.

Vi har altså de to ligningene
$${\rm Mb + O_2}\overset{k_{\rm a}}{\rightarrow} {\rm MbO_2}.$$
$${\rm MbO_2}\overset{k_{\rm d}}{\rightarrow} {\rm Mb + O_2}.$$

Ofte skrives dette som én reaksjon som kan gå i begge retninger som:
$${\rm Mb + O_2}\underset{k_{\rm d}}{\overset{k_{\rm a}}{\leftrightharpoons}} {\rm MbO_2}.$$

Vi ønsker nå igjen å skrive opp *reaksjonsraten*, altså endringen i hver konsenstrasjon. La oss igjen begynne med $[\rm Mb]$. Som før synker denne konsentrasjonen for hver oksygen som binder seg til et myoglobin, altså synker konsentrasjonen av fri myoglobin når reaksjonen utløper seg mot høyre. Derimot *dannes* det fri myoglobin hver gang et oksymyoglobin splitter seg. Altså får den deriverte to ledd, ett positivt, og ett negativt:
$$[{\rm Mb}]' = k_{\rm d}[{\rm MbO_2}] - k_{\rm a} [{\rm Mb}][{\rm O_2}].$$

[{\rm O_2}]_{i+1} &= [{\rm O_2}]_i + (k_{\rm d}[{\rm MbO}_2]_i - k_{\rm a}[{\rm Mb}]_i[{\rm O_2}]_i)\times \Delta t, \\

\end{align}
$$

## Likevekt

Når reaksjonen kan utspille seg i begge retninger vil systemet til slutt treffe en likevekt, der reaksjonen foregår like raskt i begge retninger. I plottet vårt ser vi at dette har skjedd fordi alle kurvene har flatet helt ut.

Vi ser at ved likevekt er noe av myoglobinet bundet til oksygen, mens noe av myoglobinet ikke er det. La oss prøve å finne et mål på hvor mye som er bundet.

Ved likevekt endrer ikke konsentrasjonene seg, det betyr at de deriverte er lik 0:
$$\rm [Mb]' = 0, \qquad [O_2]' = 0, \qquad [MbO_2]' = 0.$$
Om vi setter inn et av uttrykene av de deriverte får vi da ligningen
$$[{\rm Mb}]' = k_{\rm d}[{\rm MbO_2}] - k_{\rm a} [{\rm Mb}][{\rm O_2}] = 0.$$
Vi kan nå flytte rundt på de ulike størrelsene for å få likevektsuttrykket
$$\frac{[{\rm Mb}][{\rm O_2}]}{[{\rm MbO_2}]} = \frac{k_{\rm a}}{k_{\rm d}}.$$
Dette er det vi kaller *massevirkningsloven*, der man ofte skriver forholdet av de to ratekonstantene som *likevektskonstanten*, $K$, slik at vi har
$$\frac{[{\rm Mb}][{\rm O_2}]}{[{\rm MbO_2}]} = K, \qquad \mbox{der } K = \frac{k_{\rm a}}{k_{\rm d}}.$$

#### Oppgave 1e) Likevekt

Ved likevekt så skal
$$\frac{[{\rm Mb}][{\rm O_2}]}{[{\rm MbO_2}]} = K, \qquad \mbox{der } K = \frac{k_{\rm a}}{k_{\rm d}}.$$
Sjekk at dette er oppfyllt for simuleringen vår ved å lese av de tre konsentrasjonene ved slutten av simuleringen og sette inn i formlene.

Likevektskonstanten er gitt ved: $K = \frac{k_{\rm a}}{k_{\rm d}}$

### Oksygeniseringsgrad

Fra likevektsuttrykket kan vi forutsi hvor stor andel av myoglobin som har oksygen bundet til seg. Dette kaller vi *oksygeniseringsgraden* til myoglobin, vi gir denne størrelsen variabelnavnet $Y$.

Et utrykk for oksygeniseringsgraden er
$$Y = \frac{[\text{oksymyoglobin}]}{[\text{total mengde myoglobin}]},$$
med "total" mengde mener vi her både myoglobin med oksygen bundet til seg, og myoglobin uten oksygen. Da har vi
$$Y = \frac{\rm [MbO_2]}{\rm [Mb] + [MbO_2]}.$$

Oksygeniseringsgraden $Y$ er her en andel. Så om $Y$ for eksempel er lik 0.5, så er 50% av myoglobinet bundet til oksygen, mens 50% er ikke det. Om $Y$ er 0.9 så har 90% av myoglobinet oksygen bundet til seg.

Ved likevekt så kan vi finne et fint uttrykk for $Y$. Fra massevirkningsloven har vi at
$$\frac{[{\rm Mb}][{\rm O_2}]}{[{\rm MbO}_2]} = K,$$
om vi løser denne for $[{\rm MbO}_2]$ får vi:
$$[{\rm MbO}_2] = \frac{[{\rm Mb}][{\rm O_2}]}{K}.$$
Vi setter nå inn dette uttrykket i ligningen for $Y$:
$$Y = \frac{\frac{[{\rm Mb}][{\rm O_2}]}{K}}{[{\rm Mb}] + \frac{[{\rm Mb}][{\rm O_2}]}{K}}.$$
Vi kan nå stryke $\rm [Mb]$ fullstendig, fordi den finner vi i begge ledd i teller, og i nevner:
$$Y = \frac{\frac{[{\rm O_2}]}{K}}{1 + \frac{[{\rm O_2}]}{K}}.$$
Vi utvider nå brøken med $K$ for å gjøre ting penere, dvs, ganger både teller og nevner med $K$:
$$Y = \frac{[{\rm O_2}]}{[{\rm O}_2] + K}.$$

Vi har nå et pent uttrykk for hvordan oksygeniseringsgraden av myoglobin øker etterhvert som mengden oksygen økes. Vi kan plotte denne som følger:

```python
from pylab import *

k_a = 14 # Assoisasjonsraten
k_d = 12 # Dissosiasjonsraten
K = k_a/k_d # Likevektskonstanten

O2 = linspace(0, 100, 1001) # µM
Y = O2/(O2 + K)

plot(O2, Y)
xlabel('Oksygenkonsentrasjon [µM]')
ylabel('Oksygeniseringsgrad av myoglobin, Y')
xlim(0, 100)
ylim(0, 1)
show()
```

### Oksygenbinding til hemoglobin

Som tidligere begynner vi med å se på hvordan oksygen binder seg til hemoglobin. Hemoglobin har 4 domener der oksygen kan binde seg. Dermed kan et gitt hemoglobin-protein ha 0, 1, 2, 3 eller 4 oksygengass-molekyler bundet til seg. Det er altså totalt 5 ulike tilstander hemoglobinet kan være i. Vi kan tegne opp situasjonen som følger:



Fordi et hemoglobin har 4 mulige bindinger kunne vi tenkt oss at hemoglobin vil oppføre seg helt likt som myoglobin, men ved 4 ganger høyere konsentrasjon. Altså, at vi bare kan modellere antall tilgjenlige bindinger. Dette ville vært en mulig snarvei dersom binding av oksygen til hemoglobinets fire bindinger hadde vært helt uavhenig av hverandre.

Om man gjennomfører eksperimenter for å se hvordan oksygen binder seg til hemoglobin ser vi derimot at dette *ikke* er tilfellet. Tvert om, det viser seg at i praksis vil hemoglobin stort sett enten ha 0 oksygen-atomer bundet til seg, eller alle 4. Man ser kun et relativt lite antall hemoglobin med 1-3 oksygenatomer bunder til seg.

Dette skjer fordi hemoglobin-proteinet har en slik struktur at dersom ett oksygen-atom binder seg, så blir det langt lettere for flere oksygen-atomer å binde seg. Vi kaller dette fenomenet for *kooperativ binding*, fordi de 4 bindingsstedene *samarbeider* om å binde oksygen.

For å modellere hemoglobin som binder oksygen til seg kan vi derfor formulere det som den kjemiske reaksjonen
$$\rm Hb + 4O_2 \leftrightharpoons Hb(O_2)_4.$$
Altså at fire oksygenatomer binder *samtidig* til hemoglobinet. Som tidligere har vi også to ratekonstanter her, en for hver retning av reaksjonen.

### De deriverte

For å simulere binding av oksygen til hemoglobin må vi nå sette opp de deriverte. Disse blir nå litt annerledes enn for myoglobin, ettersom at 4 oksygenmolekyler binder seg samtidig. Vi får:
$$[{\rm Hb}]' = -k_{\rm a}[{\rm Hb}][{\rm O}_2]^4 + k_{\rm d}[{\rm Hb(O}_2)_4].$$
Her blir raten avhengig av oksygenkonsentrasjonen i 4, fordi fire oksygenmolekyler må binde samtidig.

Fordi det går med 4 oksygenatomer for hver reaksjon som skjer, så blir konsentrasjonsendring av oksygen til
$$[{\rm O_2}]' = -4k_{\rm a}[{\rm Hb}][{\rm O}_2]^4 + 4k_{\rm d}[{\rm Hb(O}_2)_4].$$

Til slutt har vi 
$$[{\rm Hb(O_2)_4}]' = k_{\rm a}[{\rm Hb}][{\rm O}_2]^4 - k_{\rm d}[{\rm Hb(O}_2)_4].$$

### Likevekten

Igjen vil reaksjonen inntreffe en likevekt, der ingenting endrer seg. Som tidligere kan vi finne denne ved å sette en av de deriverte lik 0:
$$[{\rm Hb}]' = -k_{\rm a}[{\rm Hb}][{\rm O}_2]^4 + k_{\rm d}[{\rm Hb(O}_2)_4] = 0.$$
Som gir likevektsligningen
$$\frac{\rm [Hb][O_2]^4}{\rm [Hb(O_2)_4]} = K.$$

Vi får altså en tilsvarende likevekt som hos myoglobin, men fordi de 4 hemegruppene til hemoglobin samarbeider og binder oksygen samtidig vil avhengigheten av oksygenkonsentrasjonen være opphøyd i 4.

### Oksygeniseringsgrad

Vi ønsker nå å finne oksygeniseringsgraden til hemoglobin slik som vi gjorde for myoglobin. Da starter vi med uttrykket
$$Y = \frac{\rm [Hb(O_2)_4]}{\rm [Hb] + [Hb(O_2)_4]}.$$

Vi gjennomfører nå akkurat disse samme stegene som sist. Vi løser likevektsligningen for $\rm [Hb(O_2)_4]$:
$$[{\rm Hb(O_2)_4}] = \frac{\rm [Hb][O_2]^4}{K}.$$

Setter vi inn dette i uttrykket for $Y$ og rydder opp får vi da
$$Y = \frac{[{\rm O}_2]^4}{[{\rm O}_2]^4 + K}.$$

## Sammenligne oksygenbinding av myoglobin og hemoglobin

Vi har nå funnet uttrykk for hvordan oksygen og hemoglobin binder oksygen ved likevekt. Disse to uttrykkene var
$$Y_{\rm myoglobin} =\frac{[{\rm O}_2]}{[{\rm O}_2] + K_{\rm myo}}, \qquad Y_{\rm hemoglobin} = \frac{[{\rm O}_2]^4}{[{\rm O}_2]^4 + K_{\rm hemo}}.$$
Disse to uttrykkene er ganske like, men vi ser hemoglobinet har en eksponent som vil vise seg å være veldig viktig.

```python
from pylab import *

O2 = linspace(0, 6, 1001) # µM

# Oksygeniseringskurve Myoglobin
k_a = 14 # Assoisasjonsraten
k_d = 12 # Dissosiasjonsraten
K = k_a/k_d # Likevektskonstanten

Y = O2/(O2 + K)
plot(O2, Y, label='Myoglobin')

# Oksygeniseringskurve Hemoglobin
k_a = 66
k_d = 20
K = k_a/k_d 

Y = O2**4/(O2**4 + K)
plot(O2, Y, label='Hemoglobin')

xlabel('Oksygenkonsentrasjon [µM]')
ylabel('Oksygeniseringsgrad')
legend()
show()
```

# 3) Karbonmonoksid

Karbonmonoksid, CO, er et molekyl som på mange måter ligner på et oksygengass-molekyl. Derfor klarer karbonmonoksid å binde seg til myoglobin og hemoglobin, slik som oksygen gjør. Det viser seg at dette gjør karbonmonoksid meget giftig, og vi skal nå undersøke nærmere akkurat hvorfor karbonmonoksid er så farlig.

### Reaksjoner

Når vi puster inn karbonmonoksid i lungene vil det være reaksjoner mellom hemoglobin i blodet og oksygengass, og  samtidig reaksjoner mellom hemoglobinet og karbonmonoksid. Vi kan skrive opp disse reaksjonene som

$$\rm Hb + 4O_2 \leftrightharpoons Hb(O_2)_4.$$
$$\rm Hb + 4CO \leftrightharpoons Hb(CO)_4.$$

Disse to reaksjonene har ulike reaksjonsrater. Disse er som følger ([Kilde](https://www.ncbi.nlm.nih.gov/books/NBK153687/table/T22/)):
$$
\begin{align}
k_{\rm a, O_2} &= 66 {\rm\ µm^{-1}\ s^{-1}}, \\
k_{\rm d, O_2} &= 20 {\rm\ µm^{-1}\ s^{-1}}, \\
k_{\rm a, CO} &= 6 {\rm\ µm^{-1}\ s^{-1}}, \\
k_{\rm d, CO} &= 0.008 {\rm\ µm^{-1}\ s^{-1}}, \\
\end{align}
$$



### De deriverte

Det er nå snakk om totalt 5 ulike konsentrasjoner. Å sette opp de deriverte blir derfor litt av en jobb, og vi må holde tunga rett i munnen!

Vi begynner med hemoglobin. Denne inngår i to ulike reaksjoner, så det vil være to negative ledd, og to positive ledd. Disse blir som følger:
$$
\begin{align}
[{\rm Hb}]' =& -k_{\rm a,O_2}[{\rm Hb}][{\rm O}_2]^4 + k_{\rm d,O_2}[{\rm Hb(O}_2)_4] \\
       & -k_{\rm a,CO}[{\rm Hb}][{\rm CO}]^4 + k_{\rm d,CO}[{\rm Hb(CO)}_4]
\end{align}
$$

For oksygen og karbonmonoksid blir det enklere, for hver av disse inngår i kun én av de to reaksjonene:
$$
[{\rm O_2}]' = -4k_{\rm a,O_2}[{\rm Hb}][{\rm O}_2]^4 + 4k_{\rm d,O_2}[{\rm Hb(O}_2)_4] 
$$
og
$$[{\rm CO}]' = -4k_{\rm a,CO}[{\rm Hb}][{\rm CO}]^4 + 4k_{\rm d,CO}[{\rm Hb(CO)}_4]$$

Til slutt har vi hemoglobin bunded med oksygen Hb(O$_2)_4$ og bundet med karbonmonoksid Hb(CO)$_4$, disse inngår også kun i én reaksjon hver:
$$
[{\rm Hb(O_2)_4}]' = k_{\rm a,O_2}[{\rm Hb}][{\rm O}_2]^4 - k_{\rm d,O_2}[{\rm Hb(O}_2)_4] 
$$
og
$$[{\rm Hb(CO)_4}]' = k_{\rm a,CO}[{\rm Hb}][{\rm CO}]^4 - k_{\rm d,CO}[{\rm Hb(CO)}_4]$$



```python
from pylab import *

k_aO2 = 66 # 1/(µM s)
k_dO2 = 20 # 1/s
k_aCO = 6 # 1/(µM s)
k_dCO = 0.008 # 1/s

T = 100 # Antall sekunder
dt = 0.0001 # Tidssteg
N = int(T/dt) # Antall tidssteg

# Lage tomme arrays for å lagre løsninger
t = zeros(N+1)
Hb = zeros(N+1)
O2 = zeros(N+1)
CO = zeros(N+1)
HbO2 = zeros(N+1)
HbCO = zeros(N+1)

# Initielle konsentrasjoner
Hb[0] = 5 # µM
O2[0] = 20 # µM
CO[0] = 1 # µM
HbO2[0] = 0 # µM
HbCO[0] = 0 # µM

# Regne oss frem i tid (Euler's metode)
for i in range(N):
    t[i+1] = t[i] + dt
    
    Hb[i+1] = Hb[i] + (k_dO2*HbO2[i] + k_dCO*MbCO[i] - k_aO2*Hb[i]*O2[i]**4 - k_aCO*Hb[i]*CO[i]**4)*dt
    
    O2[i+1] = O2[i] + (k_dO2*HbO2[i] - k_aO2*Hb[i]*O2[i]**4)*dt
    CO[i+1] = CO[i] + (k_dCO*HbCO[i] - k_aCO*Hb[i]*CO[i]**4)*dt
    
    HbO2[i+1] = HbO2[i] + (k_aO2*Hb[i]*O2[i] - k_dO2*HbO2[i])*dt
    HbCO[i+1] = HbCO[i] + (k_aCO*Hb[i]*CO[i] - k_dCO*HbCO[i])*dt
    
plot(t, Hb, label='Hb')
plot(t, O2, label='O2')
plot(t, CO, label='CO')
plot(t, HbO2, label='HbO2')
plot(t, HbCO, label='HbCO')
xlabel('Tid (ms)')
ylabel('Konsentrasjoner (µM)')
legend()
show()
```
